```python
#generating dataset,
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(1,10) 
Y = [np.random.normal(3*x+3, 1.2) for x in X]
plt.scatter(X,Y)
plt.plot(X,3*X+3) 
```

```python
X = (X - np.mean(X))/np.std(X) #standardization
Y = (Y - np.mean(Y))/np.std(Y) #standardization

data = np.stack([X,Y])
cov_matrix = np.cov(data)


w,v = np.linalg.eig(cov_matrix)
v = np.transpose(v)
v = v[np.argsort(-w)]
w = w[np.argsort(-w)]
data_after_pca = np.matmul(v,data)



plt.ylim(np.mean(data_after_pca[1])-5, np.mean(data_after_pca[1])+5)

plt.scatter(data_after_pca[0], data_after_pca[1])

```
